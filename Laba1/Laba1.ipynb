## Текстовое описание датасета:
------
В качестве набора данных мы будем использовать набор данных, [содержащий информацию о различных характеристиках автомобилей.](https://www.kaggle.com/toramky/automobile-dataset)

  wheel-base  | Длина колесной базы.
  length  | Длина автомобиля.
  width  | Ширина автомобиля.
  height  | Высота автомобиля.
  engine-type  | Тип двигателя.
  num-of-cylinders  | Число цилиндров в двигателе.
  engine-size  | Рабочий объем двигателя.
  bore  | Диамер поршня.
  stroke  | Ход поршня.
  compression-ratio  | Степень сжатия двигателя.
  horsepower  | Мощность двигателя.
  peak-rpm  | Максимально допустимое число оборотов.
  city-mpg  | Дальность хода миль\галлон в городе.
  highway-mpg  | Дальность хода миль\галлон за городом.
  price  | Стоимость автомобиля.

  
  ## Загрузка датасета:
  -------
  Загрузим данные из датасета с помощью библиотеки pandas.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set(style="ticks")

# Загрузка данных
data = pd.read_csv('data/Automobile_data.csv', sep=';')
# Вывод первых 5 строк из датасета
data.head()
```

```python
# Выведем в адекватном виде:
print('Всего строк: {}'.format(data.shape[0]))
print('Всего столбцов: {}'.format(data.shape[1]))
```

```python
# Выведем список колонок с типом данных
data.dtypes
```

```python
# Проверим датасет на наличие пустых значений
for col in data.columns:
    print('{} - {}'.format(col, data[data[col].isnull()].shape[0]))
```

```python
# Определим уникальные значения для 'price'
len(data['price'].unique())
```

  ## Визуальное исследование датасета:
  -------
  ### Диаграмма рассеивания
  --------
Позволяет построить распределение двух колонок данных и визуально обнаружить наличие зависимости.

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.scatterplot(ax=ax, x='horsepower', y='price', data=data)

# Так как число возможных значений большое, с графика уберем подписи
ax.set_yticklabels([])
ax.set_xticklabels([])
```

На диаграмме видно, что между мощностью автомобиля и его стоимостью есть зависимость.

### Гистограмма
--------
Позволяет оценить плотность вероятности распределения данных.

```python
fig, ax = plt.subplots(figsize=(20,20)) 
sns.distplot(data['price'])
```

### Joinplot
------
Позволяет построить комбинацию диаграммы рассеивания и гистограммы.

```python
sns.jointplot(x='horsepower', y='price', data=data)
```

```python
sns.jointplot(x='horsepower', y='price', data=data, kind="hex")
```

```python
sns.jointplot(x='engine-size', y='price', data=data, kind="kde")
```

### Парные диаграммы
---------
Комбинация гистограмм и диаграмм рассеивания для всего набора данных.

Выводится матрица графиков. На пересечении строки и столбца, которые соответстуют двум показателям, строится диаграмма рассеивания. В главной диагонали матрицы строятся гистограммы распределения соответствующих показателей.

```python
sns.pairplot(data)
```

### Ящик с усами
--------
Отображает одномерное распределение вероятности.

```python
sns.boxplot(x=data['engine-size'])
```

```python
# Распределение параметра engine-size сгруппированные по price.
sns.boxplot(x='num-of-cylinders', y='price', data=data)
```

### Скрипичная диаграмма
--------
Ядерная оценка плотности является задачей сглаживания данных, когда делается заключение о совокупности, основываясь на конечных выборках данных.

```python
sns.violinplot(x=data['price'])
```

```python
fig, ax = plt.subplots(2, 1, figsize=(15,10))
sns.violinplot(ax=ax[0], x=data['price'])
sns.distplot(data['price'], ax=ax[1])
```

Из приведенных графиков видно, что violinplot действительно показывает распределение плотности.

```python
sns.violinplot(x='num-of-cylinders', y='price', data=data)
```

### Информация о корреляции признаков
------

```python
data.corr()
```

```python
sns.heatmap(data.corr())
```

```python
# Вывод значений в ячейках
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(data.corr(), annot=True, fmt='.3f', ax=ax)
```

```python
corr = data.corr()
m = (corr.mask(np.eye(len(corr), dtype=bool)).abs() > 0.6).any()
raw = corr.loc[m, m]
raw
```

```python
# Треугольный вариант матрицы
mask = np.zeros_like(raw, dtype=np.bool)
# чтобы оставить нижнюю часть матрицы
# mask[np.triu_indices_from(mask)] = True
# чтобы оставить верхнюю часть матрицы
mask[np.tril_indices_from(mask)] = True
sns.heatmap(raw, mask=mask, annot=True, fmt='.3f')
```
